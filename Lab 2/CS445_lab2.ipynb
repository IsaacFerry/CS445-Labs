# Introduction
## Dataset: Differentiated Thyroid Cancer Recurrence

The Differentiated Thyroid Cancer Recurrence dataset was part of a research in the field of AI and Medicine. Each instance in the dataset represents an individual patient. 

This dataset focuses on patients with thyroid nodules and potential thyroid cancer diagnoses. It includes demographic information, clinical features, and pathological results that are used for assessing thyroid cancer risks and treatment outcomes.

### Features
This data set contains 13 clinicopathologic features aiming to predict the recurrence of well-differentiated thyroid cancer. The data set was collected over a duration of 15 years and each patient was followed for at least 10 years.

### Purpose
This data can be used to analyze correlations between clinical variables


```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the the thyroid dataset
dataset = pd.read_csv('Thyroid_Diff.csv')

# Display a preview of the first five entries 
dataset.head(5)

# Set a consistent plot style
sns.set(style="whitegrid")
```

# Visualization 1: Stage Distribution

```python
# Count plot showing how many patients fall into each cancer stage
fig1 = plt.figure(figsize=(8, 6))
sns.countplot(x='Stage', data=dataset)
plt.title("<Isaac> + Distribution of Cancer Stages")
plt.xlabel("Cancer Stage")
plt.ylabel("Number of Patients")
plt.tight_layout()

# Save plot as an image
fig1.savefig('Isaac_DistributionOfCancerStages')
```

- **Purpose**: Observe which cancer stages are most common
- **Insight**: Most patients are in the early stage (Stage I), indicating early detection
- **What can we learn**: Early diagnosis is common in this study group, possibly due to regular screening. 

# Visualization 2: Age by Risk Level

```python
# Box plot showing age distribution for each risk group
fig2 = plt.figure(figsize=(8, 6))
sns.boxplot(x='Risk', y='Age', data=dataset)
plt.title("<Isaac> + Age Distribution by Risk Level")
plt.xlabel('Risk of cancer')
plt.ylabel('Age of patients')
plt.show()

# Save plot as an image
fig2.savefig('Isaac_AgeDistributionByRiskLevel')
```

- **Purpose**: Compare the age range across Low, Intermediate, and High risk patients.
- **Insight**: The median age is higher for Intermediate and High risk groups.
- **What can we learn**: Higher risk classifications tend to be associated with older individuals.

# Visualization 3: Pathology Type by Gender

```python
# Count plot showing how pathology types differ by gender
fig3 = plt.figure(figsize=(10, 6))
sns.countplot(x='Pathology', hue='Gender', data=dataset)
plt.title("<Isaac> + Pathology Type by Gender")
plt.xlabel("Pathology Type")
plt.ylabel("Number of Patients")
plt.legend(title="Gender")
plt.show()

# Save plot as an image
fig3.savefig('Isaac_PathologyTypeByGender')
```

- **Purpose**: Show different thyroid pathology types split by gender.
- **Insight**: Papillary is most common and affects Females more than Males.
- **What we can learn**: There does not seem to be a strong gender bias in Hurthel cell from this data set.

# Visualization 4: Correlation Heatmap

```python
# Encode categorical variables
encoded_df = dataset.copy()
for col in encoded_df.columns:
    if encoded_df[col].dtype == 'object':
        encoded_df[col] = encoded_df[col].astype('category').cat.codes

fig4 = plt.figure(figsize=(14, 10))
sns.heatmap(encoded_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("<Isaac> + Correlation Heatmap of Features")
plt.show()

# Save plot as an image
fig4.savefig('Isaac_CorrelationHeatmapOfFeatures')
```

- **Purpose**: Show a matrix of pairwise correlations between all encoded categorical and numerical features.
- **Insight**: Pathology has little to no correlation to other features in the dataset
- **What we can learn**: This visualization helps identify which features may be predictive or redundant in modeling

**Additional information** The grid shows a correlation ranging from -1 to +1 and quantifies a linear relationship between two features.

- +1.00 = Perfect positive correlation (as one feature increases the other increases)
- 0.00 = No correlation
- -1.00 = Perfect negative correlation (as one feature increases the other decreases)

# Visualization 5: Violin Plot of Age by T Stage

```python
# Violin plot showing the age distribution shape by tumor stage
fig5 = plt.figure(figsize=(10, 6))
sns.violinplot(x='T', y='Age', data=dataset)
plt.title("<Isaac> + Age Distribution by Tumor Stage")
plt.xlabel('Different Tumor stages/sizes')
plt.ylabel('Age of patients in dataset')
plt.show()

# Save plot as an image
fig5.savefig('Isaac_AgeDistributionByTumorStage')
```

- **Purpose**: Show the density and range of patient ages across different tumor (T) stages.
- **Insight**: Younger patients tend to have small tumors.
- **What we can learn**: Later tumor stages seem to occur in a broader age range.
